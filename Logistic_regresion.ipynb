```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Folder containing the .npy files (S{patient_id}_raw_ungrouped.npy)
data_folder = '/Users/aidantang/Desktop/Gesture_recog/raw_dataV2'
patient_ids = range(1, 12)

# If you grouped your stimuli into 4 classes (1..4), set total_classes=4.
# If you have more classes, adjust accordingly.
total_classes = 7

results = []  # Will store dicts of patient results

def build_logistic_regression(input_dim, num_classes):
    """
    Builds a single-layer logistic regression model:
      - A Dense layer with 'num_classes' outputs + softmax activation.
    """
    model = Sequential([
        Dense(num_classes, activation='softmax', input_shape=(input_dim,))
    ])
    return model

for patient_id in patient_ids:
    file_path = os.path.join(data_folder, f'S{patient_id}_raw_ungrouped.npy')
    if not os.path.exists(file_path):
        print(f"File not found for patient {patient_id}: {file_path}")
        continue
    
    print(f"\nProcessing Patient {patient_id}")
    segments = np.load(file_path)
    print("Loaded data shape:", segments.shape)
    
    # If the file is shaped (num_segments, 400, 17):
    # Columns (metadata + EMG):
    #   0: Forearm_Percentage
    #   1: Years_Since_Amputation
    #   2: Myoelectric_Use
    #   3: Stimulus (grouped)
    #   4: Repetition
    #   5..16: EMG channels (12 total)
    
    if segments.shape[0] == 0:
        print(f"No segments found for patient {patient_id}. Skipping.")
        continue
    
    # Extract metadata from the first segment (for reference)
    forearm_percentage = segments[0, 0, 0]
    years_since_amputation = segments[0, 0, 1]
    myoelectric_use = segments[0, 0, 2]
    
    # Stimulus and Repetition are stored in columns 3 and 4 of row 0 for each segment
    y_all = segments[:, 0, 3]     # grouped stimulus
    reps_all = segments[:, 0, 4]  # repetition
    
    # Drop segments with stimulus == 0
    valid_idx = np.where(y_all != 0)[0]
    segments = segments[valid_idx]
    y_all = y_all[valid_idx]
    reps_all = reps_all[valid_idx]
    print("After dropping stimulus=0, segments shape:", segments.shape)
    
    # Extract EMG channels from columns 5..16 => shape: (num_segments, 400, 12)
    # We flatten each segment’s 400x12 into 4800 features for logistic regression
    # OR we can do an average or some feature extraction. 
    # If you want raw flattening, do:
    X_all = segments[:, :, 5:17].reshape(segments.shape[0], -1)
    
    # Train/Val/Test split by repetition
    train_idx = np.where(np.isin(reps_all, [1, 3, 5]))[0]
    val_idx   = np.where(np.isin(reps_all, [2, 4]))[0]
    test_idx  = np.where(reps_all == 6)[0]
    
    X_train = X_all[train_idx]
    X_val   = X_all[val_idx]
    X_test  = X_all[test_idx]
    
    y_train = y_all[train_idx]
    y_val   = y_all[val_idx]
    y_test  = y_all[test_idx]
    
    print("Train X shape:", X_train.shape)
    print("Val   X shape:", X_val.shape)
    print("Test  X shape:", X_test.shape)
    
    # Standardize features
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_val   = scaler.transform(X_val)
    X_test  = scaler.transform(X_test)
    
    # Convert labels to one-hot (assuming grouped stimulus is in {1..4})
    y_train_cat = to_categorical(y_train - 1, num_classes=total_classes)
    y_val_cat   = to_categorical(y_val - 1, num_classes=total_classes)
    y_test_cat  = to_categorical(y_test - 1, num_classes=total_classes)
    
    # Build logistic regression model (single Dense with softmax)
    model = build_logistic_regression(input_dim=X_train.shape[1], num_classes=total_classes)
    
    # Initial compile with higher LR
    model.compile(optimizer=Adam(learning_rate=0.1), loss='categorical_crossentropy', metrics=['accuracy'])
    print("Training on Train Data...")
    model.fit(X_train, y_train_cat, epochs=10, batch_size=32, validation_data=(X_val, y_val_cat), verbose=1)
    
    # Fine-tune on validation data with lower LR
    print("Fine-Tuning on Validation Data...")
    model.compile(optimizer=Adam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(X_val, y_val_cat, epochs=5, batch_size=32, verbose=1)
    
    # Evaluate on test data
    test_loss, test_accuracy = model.evaluate(X_test, y_test_cat, verbose=0)
    print(f"Test Accuracy for Patient {patient_id}: {test_accuracy:.4f}")
    
    # Predictions & confusion matrix
    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_true = np.argmax(y_test_cat, axis=1)
    
    display_labels = list(range(1, total_classes + 1))  # e.g., [1,2,3,4]
    cm = confusion_matrix(y_true, y_pred, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=display_labels)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(cmap='viridis', values_format=".2f", ax=ax)
    plt.title(f"Patient {patient_id} - Confusion Matrix (LogReg)")
    plt.show()
    
    # Store results
    results.append({
        'PatientID': patient_id,
        'TestAccuracy': test_accuracy,
        'Years_Since_Amputation': years_since_amputation,
        'Forearm_Percentage': forearm_percentage,
        'Myoelectric_Use': myoelectric_use
    })

# Print final results
print("\nFinal Results:")
for r in results:
    print(r)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
print("Columns:", results_df.columns.tolist())

# Helper function to get the correct column name
def get_column(name_options, df):
    for name in name_options:
        if name in df.columns:
            return name
    raise KeyError(f"None of the column names {name_options} found.")

# Identify columns in your DataFrame
years_col = get_column(['Years_Since_Amputation', 'YearsSinceAmputation'], results_df)
forearm_col = get_column(['Forearm_Percentage', 'ForearmPercentage'], results_df)
myo_col = get_column(['Myoelectric_Use', 'MyoelectricUse'], results_df)

# 1) Accuracy vs. Years Since Amputation
plt.figure(figsize=(7, 5))
plt.scatter(results_df[years_col], results_df['TestAccuracy'], c='blue', label='Accuracy')
coeffs = np.polyfit(results_df[years_col], results_df['TestAccuracy'], deg=1)
line = np.poly1d(coeffs)
x_vals = np.linspace(results_df[years_col].min(), results_df[years_col].max(), 100)
plt.plot(x_vals, line(x_vals), color='blue', label='Best Fit')
plt.xlabel('Years Since Amputation')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Years Since Amputation (LogReg)')
plt.grid(True)
plt.legend()
plt.show()

# 2) Accuracy vs. Forearm Percentage
plt.figure(figsize=(7, 5))
plt.scatter(results_df[forearm_col], results_df['TestAccuracy'], c='green', label='Accuracy')
coeffs = np.polyfit(results_df[forearm_col], results_df['TestAccuracy'], deg=1)
line = np.poly1d(coeffs)
x_vals = np.linspace(results_df[forearm_col].min(), results_df[forearm_col].max(), 100)
plt.plot(x_vals, line(x_vals), color='green', label='Best Fit')
plt.xlabel('Forearm Percentage')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Forearm Percentage (LogReg)')
plt.grid(True)
plt.legend()
plt.show()

# 3) Accuracy vs. Myoelectric Use
plt.figure(figsize=(7, 5))
plt.scatter(results_df[myo_col], results_df['TestAccuracy'], c='red', label='Accuracy')
coeffs = np.polyfit(results_df[myo_col], results_df['TestAccuracy'], deg=1)
line = np.poly1d(coeffs)
x_vals = np.linspace(results_df[myo_col].min(), results_df[myo_col].max(), 100)
plt.plot(x_vals, line(x_vals), color='red', label='Best Fit')
plt.xlabel('Myoelectric Use')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Myoelectric Use (LogReg)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import scipy.stats as stats
import statsmodels.api as sm

# Statistical analysis function
def analyze_relationship(x, y, variable_name):
    # Pearson correlation
    r, p_value_corr = stats.pearsonr(x, y)
    
    # Linear Regression Analysis
    X = sm.add_constant(x)  # Add constant for intercept
    model = sm.OLS(y, X).fit()
    p_value_reg = model.pvalues[1]  # p-value for the predictor variable
    
    print(f"\nAnalysis for {variable_name}:")
    print(f"Pearson Correlation: r = {r:.4f}, p-value = {p_value_corr:.4f}")
    print(f"Linear Regression: Coefficient = {model.params[1]:.4f}, p-value = {p_value_reg:.4f}")
    
    if p_value_corr < 0.05:
        print(f"→ The correlation between {variable_name} and Accuracy is statistically significant (p < 0.05).")
    else:
        print(f"→ No significant correlation found (p >= 0.05).")
    
    if p_value_reg < 0.05:
        print(f"→ The regression model suggests a significant effect of {variable_name} on Accuracy (p < 0.05).")
    else:
        print(f"→ No significant effect found in regression (p >= 0.05).")

# Assuming results_df is your DataFrame with columns: 'Years_Since_Amputation', 
# 'Forearm_Percentage', 'Myoelectric_Use', and 'TestAccuracy'
analyze_relationship(results_df['Years_Since_Amputation'], results_df['TestAccuracy'], "Years Since Amputation")
analyze_relationship(results_df['Forearm_Percentage'], results_df['TestAccuracy'], "Forearm Percentage")
analyze_relationship(results_df['Myoelectric_Use'], results_df['TestAccuracy'], "Myoelectric Use")
```
